This is a notebook manipulating data frames and series using **Pandas** ***texte en gras***

```python
import pandas as pd
#creating our data from dictionnary
my_data = {"name": ["john", "emily", "kevin", "grace", "michael", "felicity"],
           "last name": ["doe", "stewart", "sanson", "budd", "brown", "du pain"],
           "age": [36, 24, 17, 14, 46, 28],
           "occupation": ["salaryman", "college student", "high school student", "middle school student", "policeman", "fashion designer"]
           }
data=pd.DataFrame(my_data)
data #showing our data
```

```python
data.tail(1)#showing the last row of our df
```
